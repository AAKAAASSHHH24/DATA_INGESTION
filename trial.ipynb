```python
import os
from PyPDF2 import PdfFileReader, PdfFileWriter

# Open the PDF file
input_path = 'data cleaning/Vendor Catalogues/Krannich Solar East LLC/PROD_KRUSW_Product_Catalogue_[21064]_2023-05-20.pdf'
input_pdf = PdfFileReader(open(input_path, 'rb'))

# Loop through all pages and save each page as a separate PDF file
for page_num in range(input_pdf.getNumPages()):
    # Create a new PDF writer object and add the current page to it
    output_pdf = PdfFileWriter()
    output_pdf.addPage(input_pdf.getPage(page_num))
    
    # Write the output PDF to a file with the page number in the filename
    output_filename = f'{os.path.splitext(input_path)[0]}_{page_num+1}.pdf'
    with open(output_filename, 'wb') as output_file:
        output_pdf.write(output_file)
```
