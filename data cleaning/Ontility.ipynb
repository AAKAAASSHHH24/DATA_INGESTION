```python
import pandas as pd
df_from_excel = pd.read_excel('/content/Ontility.xlsx')
```

```python
df_from_excel
```

```python
df_from_excel.info()
```

```python
df_from_excel = df_from_excel.rename(columns={'Part Number':'Item','Quantity':'Availability'})
```

```python
# Create a description column
df_from_excel['Description'] = (
    'Manufacturer: ' + df_from_excel['Manufacturer'] + ' ,' +
    'Dimensions (mm): ' + df_from_excel['Dimensions (mm)'] + ' ,' +
    'Wattage: ' + df_from_excel['Wattage'].astype(str) + ' ,' +
    'Total Watts: ' + df_from_excel['Total Watts'].astype(str) + ' ,' +
    'Bifacial: ' + df_from_excel['Bifacial'].astype(str) + ' ,' +
    'Type: ' + df_from_excel['Type'].astype(str) + ' ,' +
    'Cell: ' + df_from_excel['Cell'].astype(str) + ' ,' +
    'Pallet Size: ' + df_from_excel['Pallet Size'].astype(str) + ' ,' +
    'Location: ' + df_from_excel['Location'].astype(str) + ' ,' +
    'Min Buy: ' + df_from_excel['Min Buy'].astype(str) + ' ,' +
    'Notes: ' + df_from_excel['Notes'].astype(str)
)
```

```python
df_from_excel.columns
```

```python
df_from_excel.drop(columns= ['Manufacturer', 'Wattage', 'Dimensions (mm)', 'Type', 'Cell',
       'Bifacial', 'Pallet Size', 'Total Watts', 'Location',
       'Min Buy', 'Notes'], inplace = True)
```

```python
df_from_excel['Price']= 0
```

```python
new_column_order = ['Item', 'Description', 'Price', 'Availability']
```

```python
# Reorder the columns
df_reordered = df_from_excel.reindex(columns=new_column_order)
df_reordered
```

```python
df_reordered['Availability']=  df_reordered['Availability'].astype(str) + " items are available"
```
