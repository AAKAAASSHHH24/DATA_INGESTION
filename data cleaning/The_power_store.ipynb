FINAL DATAFRAME WILL BE THE SUMMATION OF THE TWO FILTERED DFs obtained

```python
final_df = pd.concat([filtered_df2,concat_filtered_df1], ignore_index=True)
final_df
```

We need to remove the commas and dollars from the price column

```python
final_df['Price'] = final_df['Price'].astype(str)
```

```python
# Remove commas and convert to float
for index, row in final_df.iterrows():
    if ',' in row['Price']:
      final_df.at[index, 'Price'] = row['Price'].replace(',', '')
    elif '$' in row['Price']:
      final_df.at[index, 'Price'] = row['Price'].replace('$', '')
    elif 'C ' in row['Price']:
      final_df.at[index, 'Price'] = row['Price'].replace('C ', '')
    else:
      continue
```

```python
final_df['Price'] = final_df['Price'].astype(float)
```
