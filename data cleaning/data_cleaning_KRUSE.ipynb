```python
import pandas as pd
filtered_df1 = pd.read_excel('/content/KRUSE.xlsx')
```

```python
filtered_df1['Availability'] = filtered_df1['Availability'].fillna("No info about Availability. Please verify")
```

```python
for index, row in filtered_df1.iterrows():
    filtered_df1.at[index, 'Availability'] = row['Availability']+" "+"Wattage: "+ str(row['Wattage'])+ " " + "Cell Format: "+ str(row['Cell\nFormat'])+ " "+ "Pallett: "+ str(row['# Per\npallet'])
```

```python
filtered_df1['Availability'] = filtered_df1['Availability'].astype(str)
```

```python
filtered_df1['Price'] = filtered_df1['Price'].astype(str)
```

```python
for index, row in filtered_df1.iterrows():
    if ',' in row['Price']:
      filtered_df1.at[index, 'Price'] = row['Price'].replace(',', '')
    if '$' in row['Price']:
      filtered_df1.at[index, 'Price'] = row['Price'].replace('$', '')
    if 'C ' in row['Price']:
      filtered_df1.at[index, 'Price'] = row['Price'].replace('C ', '')
    else:
      continue
```

```python
for i in range(len(filtered_df1)):
  try:
    filtered_df1['Price'][i] = float(filtered_df1['Price'][i])
  except Exception as e:
    filtered_df1['Price'][i] = 0
```

```python
filtered_df1.isna().sum()
```

```python
filtered_df1.drop(columns = ['Unnamed: 0'], inplace = True)
```

```python
filtered_df1['Price'] = filtered_df1['Price'].astype(float)
```
