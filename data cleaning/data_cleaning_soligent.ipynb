```python
df_from_excel = df_from_excel.rename(columns={'Display Name': 'Item', 'Base Price': 'Price','Available':'Availability'})
```

```python
df_from_excel['Price'].mode().values[0]
```

```python
df_from_excel.info()
```

```python
df_from_excel['Price'] = df_from_excel['Price'].astype(str)
```

```python
for index, row in df_from_excel.iterrows():
    if '\xa0' in row['Price']:
      df_from_excel.at[index, 'Price'] = row['Price'].replace('\xa0', '0')
    if '$' in row['Price']:
      df_from_excel.at[index, 'Price'] = row['Price'].replace('$', '')
    if 'C ' in row['Price']:
      df_from_excel.at[index, 'Price'] = row['Price'].replace('C ', '')
    else:
      continue
```

```python
k=0
for i in range(len(df_from_excel)):
  try:
    df_from_excel['Price'][i] = float(df_from_excel['Price'][i])
  except Exception as e:
    k+=1
    print(e, df_from_excel['Price'][i] )
print (k)
```

```python
df_from_excel['Availability'] = df_from_excel['Availability'].astype(str)
```

```python
for index, row in df_from_excel.iterrows():
    df_from_excel.at[index, 'Availability'] = row['Availability']+" "+"Units are available "
```
