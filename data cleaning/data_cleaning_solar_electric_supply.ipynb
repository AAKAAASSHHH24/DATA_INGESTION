```python
for index, row in df_from_excel.iterrows():
    df_from_excel.at[index, 'DESCRIPTION'] = row['DESCRIPTION']+" "+"Wattage: "+ str(row['Wattage'])+ " " + "Cell Format: "+ str(row['Cell\nFormat'])+ " "+ "Pallett: "+ str(row['# Per\npallet'])
```

```python
df_from_excel.columns
```

```python
df_from_excel = df_from_excel.rename(columns={'ITEM':'Item','DESCRIPTION':'Description','Pallet +\nPrice\n$/Watt': 'Price'})
```

```python
df_from_excel.drop(columns = ['Wattage', 'Cell\nFormat', '# Per\npallet'], inplace = True)
```

```python
df_from_excel
```

```python
s= '$340.00 EA\n$0.85/Watt'
s.split(' EA\n$')
```

```python
for index, row in df_from_excel.iterrows():
    df_from_excel.at[index, 'Price'] = row['Price'].split(' EA\n$')[0]
    df_from_excel.at[index, 'Description'] = row['Description'] + " "+"Pricing per watt: " + row['Price'].split(' EA\n$')[1]
```

```python
for index, row in df_from_excel.iterrows():
    if ',' in row['Price']:
      df_from_excel.at[index, 'Price'] = row['Price'].replace(',', '')
    if '$' in row['Price']:
      df_from_excel.at[index, 'Price'] = row['Price'].replace('$', '')
    if 'C ' in row['Price']:
      df_from_excel.at[index, 'Price'] = row['Price'].replace('C ', '')
    else:
      continue
```

```python
df_from_excel['Price'] = df_from_excel['Price'].astype(float)
```

```python
df_from_excel.drop(columns = ['Unnamed: 0'], inplace = True)
```
