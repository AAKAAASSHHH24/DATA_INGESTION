```python
model_name = "gpt-3.5-turbo"
temperature = 0.0
model = OpenAI(model_name=model_name, temperature=temperature)
```

```python
# Define your desired data structure.
class ProductType(BaseModel):
    Product_Type : str = Field(description="Product Type of this Product")
    Product_Sub_Type : str = Field(description="Product Sub Type of this Product")  

# Set up a parser + inject instructions into the prompt template.
parser = PydanticOutputParser(pydantic_object=ProductType)
```

```python
prompt_level1 = PromptTemplate(
    template="""
        Consider yourself an expert in analysing SOLAR POWER PRODUCTS:
        \n{format_instructions} 
        You are given the Title of the product as  the following: {title}.
        You are given the description body of the same product as the following: {body}
        For generating Product_Type select from the following list: {Product_Types}.
        For generating Product_Sub_Type select from the following list: {Product_SubTypes}.
        Only include the generated answer in your response and nothing else""",
    input_variables=["title","body","Product_Types","Product_SubTypes"],
    partial_variables={"format_instructions": parser.get_format_instructions()},
)
```

```python
_input = prompt_level1.format_prompt(title= prod['Title'][0] , body = prod['Body'][0] , Product_Types = Product_Types, Product_SubTypes = Product_SubTypes )

output_level1 = model(_input.to_string())

pprint.pprint(output_level1)
```
